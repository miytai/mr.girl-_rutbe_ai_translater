```python
!pip install -U openai-whisper
!pip install sentencepiece
!pip install pydub
!pip install argostranslate
!pip install moviepy
!pip install -U pip
!pip install pyttsx3
!sudo apt-get install espeak


import whisper
import argostranslate.package
import argostranslate.translate
from pydub import AudioSegment
import torch
from moviepy.editor import *
import pyttsx3


video = VideoFileClip('/content/video.mp4')
audio = video.audio
audio.write_audiofile('/content/audio.mp3')

def trans():
  from_code = "ru"
  to_code = "en"

  # Download and install Argos Translate package
  argostranslate.package.update_package_index()
  available_packages = argostranslate.package.get_available_packages()
  package_to_install = next(
      filter(
          lambda x: x.from_code == from_code and x.to_code == to_code, available_packages
      )
  )
  argostranslate.package.install_from_path(package_to_install.download())
  file_path = "/content/transcription_tiny.txt"  # Укажите путь к вашему файлу

  with open(file_path, 'r', encoding='utf-8') as file:
      text = file.read()
  # Translate
  translatedText = argostranslate.translate.translate(text, from_code, to_code)
  print(translatedText)
  with open('output.txt', 'w', encoding='utf-8') as file:
    file.write(translatedText)
  reading(translatedText)
  return (translatedText)

def speech_recognition(model='base'):
  speech_model = whisper.load_model(model)
  result = speech_model.transcribe('/content/audio.mp3')

  with open(f'transcription_{model}.txt', "w") as file:
    file.write(result['text'])
  trans()

def reading(translatedText):
  engine = pyttsx3.init()
  # Преобразуем текст в речь
  text = translatedText
  rate = engine.getProperty('rate')   # получаем текущую скорость чтения
  engine.setProperty('rate', rate-50)
   # Установить нужную голосовую модель
  engine.save_to_file(text, 'output_translate.mp3')
  engine.runAndWait()
  video_record()


def video_record():
  video = VideoFileClip("/content/video.mp4")
  new_audio = AudioFileClip("/content/output_translate.mp3")
  # Заменяем аудиодорожку в видео и сохраняем новое видео
  video = video.set_audio(new_audio)
  video.write_videofile("/content/translate_video.mp4")



def main():
  models = {1: 'tiny', 2: 'base', 3: 'small', 4: 'medium', 5: 'large'}
  for k, v in models.items():
    print(f'{k}:{v}')
  model_choice = int(input("Choose a model (1-5): "))
  model_name = models.get(model_choice)
  if model_name:
    print(f"Starting speech recognition using model: {model_name}")
    speech_recognition(model_name)
  else:
    print("Invalid model choice. Please choose a number between 1 and 5.")

if __name__ == '__main__':
  main()
```
